```python
#1.
import datetime

def tarik_tunai(saldo, jumlah_tarik):
    if jumlah_tarik > saldo:
        print("saldo anda tidak cukup")
        return saldo
    
    if jumlah_tarik > 1000: 
        biaya = 3
    elif saldo < 100:
        biaya = jumlah_tarik * 0.03
    else:
        biaya = 0
    
    daily_limit = 1000
    current_hour = datetime.datetime.now().hour
    
    if current_hour >= 23 or (current_hour >= 0 and current_hour < 2):
        print("Maaf, layanan ATM tidak tersedia saat ini")
        return saldo
    
    if jumlah_tarik > daily_limit:
        print("Over daily limit")
        return saldo
    
    saldo -= jumlah_tarik
    saldo -= biaya
    
    return saldo

# Contoh penggunaan
saldo_awal = 500
print("Saldo awal: $", saldo_awal)

jumlah_tarik = int(input("Masukkan jumlah yang ingin ditarik: $"))
saldo_akhir = tarik_tunai(saldo_awal, jumlah_tarik)

print("Saldo akhir: $", saldo_akhir)
```

```python
#2.
def menghitung_harga_taksi(jarak, waiting_time):
    initial_flag = 0.5
    
    # Ongoing Drops
    ongoing_drops=0
    if jarak > 2:
        ongoing_drops+=1.5  
        ongoing_drops+=1.3*(jarak-2)  
    
    # Waiting Time
    waiting_tariff=0
    if waiting_time<=120:
        waiting_tariff+=0.20*waiting_time  
    else:
        waiting_tariff+=0.20*120  
        waiting_tariff+=0.15*(waiting_time-120)  
    
    # hitung Trip Rate
    trip_rate = initial_flag + ongoing_drops + waiting_tariff
    
    return trip_rate


# uji fungsi
jarak = float(input("Enter the distance covered (in miles): "))
waiting_time = int(input("Enter the waiting time (in minutes): "))

harga = menghitung_harga_taksi(jarak, waiting_time)
print("Total Trip Fare: $", harga)

    
```

```python
#4.
def hitung_faktorial(n):
    faktorial = 1
    for i in range(1, n + 1):
        faktorial *= i
    return faktorial

def hitung_permutasi(n, r):
    permutasi = hitung_faktorial(n) // hitung_faktorial(n - r)
    return permutasi

def hitung_kombinasi(n, r):
    kombinasi = hitung_faktorial(n) // (hitung_faktorial(r) * hitung_faktorial(n - r))
    return kombinasi


# ============================
# INPUT USER
# ============================
n = int(input("Masukkan nilai n: "))
r = int(input("Masukkan nilai r: "))

# Validasi sederhana
if r > n:
    print("Nilai r tidak boleh lebih besar dari n")
else:
    # ============================
    # PROSES PERHITUNGAN
    # ============================
    faktorial_n = hitung_faktorial(n)
    permutasi_nr = hitung_permutasi(n, r)
    kombinasi_nr = hitung_kombinasi(n, r)

    # ============================
    # OUTPUT
    # ============================
    print("\nHasil Perhitungan:")
    print(f"{n}! = {faktorial_n}")
    print(f"P({n},{r}) = {permutasi_nr}")
    print(f"C({n},{r}) = {kombinasi_nr}")
```

```python
# Program Kasir Sederhana (Versi Lebih Efisien & Menarik)

# ============================
# DATA PRODUK
# ============================
harga_produk = {
    "sabun": 5000,
    "shampoo": 10000,
    "odol": 8000
}

keranjang = {}  # menyimpan item yang dibeli

# ============================
# FUNGSI TAMPILKAN PRODUK
# ============================
def tampilkan_produk():
    print("\n=== DAFTAR PRODUK ===")
    for nama, harga in harga_produk.items():
        print(f"- {nama.capitalize():10} : Rp{harga:,}")
    print()

# ============================
# FUNGSI INPUT BELANJA
# ============================
def tambah_ke_keranjang():
    while True:
        tampilkan_produk()

        produk = input("Masukkan nama produk : ").lower()

        if produk not in harga_produk:
            print("❌ Produk tidak tersedia!\n")
            continue

        try:
            jumlah = int(input("Masukkan jumlah beli : "))
            if jumlah <= 0:
                print("Jumlah harus lebih dari 0!\n")
                continue
        except ValueError:
            print("Input jumlah harus angka!\n")
            continue

        # Simpan ke keranjang
        if produk in keranjang:
            keranjang[produk] += jumlah
        else:
            keranjang[produk] = jumlah

        print(f"✅ {jumlah} {produk} ditambahkan ke keranjang.")

        lanjut = input("Tambah produk lain? (y/n): ").lower()
        if lanjut == 'n':
            break

# ============================
# FUNGSI HITUNG TOTAL
# ============================
def hitung_total():
    total = 0
    print("\n=== STRUK PEMBELIAN ===")
    print("---------------------------------")

    for produk, jumlah in keranjang.items():
        harga = harga_produk[produk]
        subtotal = harga * jumlah
        total += subtotal
        print(f"{produk.capitalize():10} x{jumlah:<3} = Rp{subtotal:,}")

    print("---------------------------------")
    print(f"TOTAL BAYAR        = Rp{total:,}")
    print("---------------------------------")

# ============================
# PROGRAM UTAMA
# ============================
print("🛒 Selamat Datang di Mini Market Python")

tambah_ke_keranjang()
hitung_total()

print("\nTerima kasih telah berbelanja!")
```
